```python
import pandas as pd

def align_nc_csv(nc_dates, nc_values, csv_df, var_name, precision='D'):
    """
     NetCDF  CSV  NaN  DataFrame
    -------------------------------------------------------------------------
    nc_dates   : 1-D array-like, datetime  pd.to_datetime 
    nc_values  : 1-D array-like nc_dates 
    csv_df     : DataFrame ['date', var_name]
    var_name   : str csv_df  'd18op'
    precision  : str, 'D'=, 'M' , 'H'   .dt.floor()
    -------------------------------------------------------------------------
    return     : DataFrame(columns=['date_only', f'{var_name}_nc', f'{var_name}_csv'])
    """
    # --- 1. NetCDF    DataFrame ---
    df_nc = pd.DataFrame({
        'date_nc': pd.to_datetime(nc_dates),
        f'{var_name}_nc': nc_values
    })
    df_nc['date_only'] = df_nc['date_nc'].dt.floor(precision)

    # --- 2. CSV  date_only ---
    if 'date' not in csv_df.columns:
        raise KeyError("'csv_df'  'date'")
    if var_name not in csv_df.columns:
        raise KeyError(f"'csv_df'  '{var_name}'")

    df_csv = csv_df[['date', var_name]].rename(columns={var_name: f'{var_name}_csv'})
    df_csv['date_only'] = pd.to_datetime(df_csv['date']).dt.floor(precision)

    # --- 3.  date_only  ---
    merged = df_nc.merge(df_csv, on='date_only', how='inner')

    # --- 4.  ---
    clean = merged.dropna(subset=[f'{var_name}_nc', f'{var_name}_csv'])

    # --- 5.  ---
    return clean[['date_only', f'{var_name}_nc', f'{var_name}_csv']].reset_index(drop=True)
```
